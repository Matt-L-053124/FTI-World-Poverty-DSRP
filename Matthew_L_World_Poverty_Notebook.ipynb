```python
#POLAND


pre_91 = poland_sorted[poland_sorted["year"] <= 1990]
post_91 = poland_sorted[poland_sorted["year"] > 1990]

pre_group = pre_91["headcount_ratio_international_povline"]
post_group = post_91["headcount_ratio_international_povline"]



```

H0: The average poverty rate in Poland prior to 1991 is the same post 1991
Hu: The average poverty rate in Poland prior to 1991 is different than post 1991

```python
import math

xbar1 = pre_group.mean()
s1 = pre_group.std()
n1 = len(pre_group)

xbar2 = post_group.mean()
s2 = post_group.std()
n2 = len(post_group)

standard_error = math.sqrt((s1**2 / n1) + (s2**2 / n2))

t_stat = (xbar1 - xbar2) / standard_error

print("Pre-1991 Poverty Mean:", xbar1)
print("Post-1991 Poverty Mean:", xbar2)
print("Standard Error:", standard_error)
print("T statistic:", t_stat)
```

```python
from scipy import stats

degrees_of_freedom = n1 + n2 - 2

p_val = stats.t.sf(abs(t_stat), df=degrees_of_freedom) * 2

print("P-value: ", p_val)
```

This proves that the fall of the Soviet Union had a significant impact/effect on Poland's poverty rates
