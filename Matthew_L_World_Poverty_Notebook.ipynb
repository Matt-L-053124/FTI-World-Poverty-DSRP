```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'poverty_original.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
columns_kept = [
    'country', 'year', 'reporting_level', 'welfare_type', 'ppp_version',
    'survey_year', 'survey_comparability', 'headcount_ratio_international_povline',
    'headcount_ratio_lower_mid_income_povline', 'headcount_ratio_upper_mid_income_povline',
    'headcount_ratio_100', 'headcount_ratio_1000', 'headcount_ratio_2000',
    'headcount_ratio_3000', 'headcount_ratio_4000', 'headcount_ratio_40_median',
    'headcount_ratio_50_median', 'headcount_ratio_60_median', 'headcount_international_povline',
    'headcount_lower_mid_income_povline', 'headcount_upper_mid_income_povline',
    'headcount_100', 'headcount_1000', 'headcount_2000', 'headcount_3000', 'headcount_4000',
    'poverty_gap_index_international_povline', 'poverty_gap_index_lower_mid_income_povline',
    'poverty_gap_index_upper_mid_income_povline', 'poverty_gap_index_100',
    'poverty_gap_index_1000', 'poverty_gap_index_2000', 'poverty_gap_index_3000',
    'poverty_gap_index_4000', 'mean', 'median', 'gini', 'mld', 'polarization', 'palma_ratio'
]

df = df[columns_kept]
df
```

```python
df['reporting_level'].unique()

df[df['reporting_level'].isnull()]

```

Code below drops all NAN values of reporting_level, welfare_type, survey_year, survey_comparability, etc

```python
df = df.dropna(subset=['reporting_level', 'welfare_type', 'survey_year', 'survey_comparability', 'headcount_ratio_40_median', 'headcount_ratio_50_median', 'headcount_ratio_60_median'])

df.info()
```

```python


df[df['gini'].isnull()]
```

Dropping all code with null values in gini, mld, polarization, palma_ratio

```python
df = df.dropna(subset=['gini', 'mld', 'polarization', 'palma_ratio'])

df.info()
```

```python
countries_plot = ["United States", "India", "Vietnam", "France", "Brazil"]
bar_df = df[(df["year"] == 2018) & (df["country"].isin(countries_plot))]

plt.figure(figsize=(8,5))
plt.bar(bar_df["country"], bar_df["gini"])

plt.title("Gini Coefficient in the year 2018 in various countries")
plt.xlabel("Country")
plt.ylabel("Gini Coefficient")

plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()

```

```python
scatter = df[df['year'] == 1994]

plt.figure(figsize=(9,6))
plt.scatter(scatter['mean'], scatter['gini'], alpha=0.5)

plt.title("Mean Daily Income vs Gini Score in 1994")
plt.xlabel("Mean Income/Expenses")
plt.ylabel("Gini Coefficient")

plt.show()
```

```python
scatter = df[df['year'] == 2006]

plt.figure(figsize=(9,6))
plt.scatter(scatter['mean'], scatter['palma_ratio'], alpha=0.5)

plt.title("Mean Daily Income vs Palma Ratio in 2006")
plt.xlabel("Mean Income/Expenses")
plt.ylabel("Palma Ratio")

plt.show()
```

```python
argentina_clean = df[(df["country"] == "Argentina") & (df["ppp_version"] == 2011) & (df["reporting_level"] == "urban")]

argentina_sorted = argentina_clean.sort_values("year")

x = argentina_sorted["year"]
y = argentina_sorted["headcount_ratio_international_povline"]

plt.plot(x,y)

plt.title("Poverty Ratio Line in Argentina over Time")
plt.xlabel("Year")
plt.ylabel("Percent of Population in Poverty")

plt.show()

```
